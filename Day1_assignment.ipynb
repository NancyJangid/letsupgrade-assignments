```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#plot a line plot between a and b:
a=np.arange(40,50)
b=np.arange(50,60)
print(a)
print(b)
plt.plot(a,b, 'ro')
plt.title('line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
#Plot a line plot showing the sales trend in company 1 and 2:

days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
print("Days of the week: {}" .format(days))
print("Sales of company 1: {}" .format(sales_1))
print("Sales of company 2: {}" .format(sales_2))
```

```python
plt.plot(days, sales_1,'go--', label='sales_1',marker='o')
plt.plot(days, sales_2, 'r*--', label='sales_2',marker='*')
plt.title('Sales distribution',color='red')
plt.xlabel('days of the week',color='b')
plt.ylabel('sales per day',color='b')
plt.legend(loc='lower right')
```

```python
# Create a 3 by 3 subplots:
# multiple plots
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]

plt.subplot(3,3,1)   # 3 row 3 col 1st segment
plt.plot(x,y1,'r--')
plt.subplot(3,3,2)   # 3 row 3 col 2nd segment
plt.plot(x,y2,'g*--')
plt.subplot(3,3,3)   # 3 row 3 col 3rd segment
plt.plot(x,y3,'bo')
plt.subplot(3,3,4)   # 3 row 3 col 4th segment
plt.plot(x,y4,'p:')
plt.subplot(3,3,5)   # 3 row 3 col 5th segment
plt.plot(x,y5,'g:')
```
